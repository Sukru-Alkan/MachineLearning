```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Dosya yolu
file_path = r"C:\Users\SukruAlkan\Desktop\football-player-position.csv"

# Veriyi oku
df = pd.read_csv(file_path)

# Hedef stun dndaki tm saysal stunlar se
features = df.drop(columns=['class'])

# Normalizasyon
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(features)

# Tekrar DataFrame'e evir
normalized_df = pd.DataFrame(normalized_data, columns=features.columns)

# Hedef stunu tekrar ekle
normalized_df['class'] = df['class']

# lk 5 satr gster
print(normalized_df.head())
```

```python


# 'class' gibi hedef deikeni ayr (eer saysalsa ama kategorikse, ilem dnda tut)
target_col = 'class'
numeric_cols = df.select_dtypes(include='number').columns.tolist()
if target_col in numeric_cols:
    numeric_cols.remove(target_col)

# Grafik izme fonksiyonu
def draw_boxplots(df, cols, title_prefix=""):
    plt.figure(figsize=(15, 5))
    for i, col in enumerate(cols, 1):
        plt.subplot(1, len(cols), i)
        sns.boxplot(y=df[col], color='skyblue')
        plt.title(f"{title_prefix} {col}")
        plt.tight_layout()
    plt.show()

# Aykr deerleri temizleme (IQR)
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]

# 1. Temizlemeden nce grafik
draw_boxplots(df, numeric_cols[:4], "nce")  # lk 4 stunu rnek olarak iz

# 2. Aykrlar temizle
for col in numeric_cols:
    df = remove_outliers_iqr(df, col)

# 3. Temizledikten sonra grafik
draw_boxplots(df, numeric_cols[:4], "Sonra")  # Ayn stunlar iz

# 4. Temizlenmi verinin boyutu
print(f"\n Temizlenmi veri boyutu: {df.shape}")
```
